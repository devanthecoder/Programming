```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
streak = lambda y, t: y**(1+2*t - np.log(y))
# stream = lambda y, t: y**(1+2*t)
path = lambda y: y**(1+np.log(y))
y = np.linspace(0, 80, 100000)
t = np.linspace(0, 1, 3)
for i in range(len(t)):
    # plt.plot(stream(y, t[i]), y, label = f'Streamline at t = {np.round(t[i], 2)}')
    plt.plot(streak(y, t[i]), y, label = f'Streakline at t = {np.round(t[i], 2)}')
plt.plot(1, 1, label = f'Point(1, 1)',marker='o')
plt.ylim([0, 10])
plt.xlim([0, 10])
plt.legend()
plt.show()
```
