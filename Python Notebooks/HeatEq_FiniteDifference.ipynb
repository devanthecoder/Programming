```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

Heat Equation

1. Length L
2. Initial temperature profile, T(x) = sin(pi*x/L)
3. Boundary conditions

u(x, t)
u(0, t) = 0 = u(L, t)

```python
#length
L = 1.0

#spacial divisions
nx = 20
x = np.linspace(0, L, nx+1)
# print(x)
dx = L/nx

#timestep
dt = 0.001

#iteration
nt = 1000

t = np.linspace(0, nt*dt, nt+1)
# print(t)

#kappa
k = 1.0

#Fourier number
F = (k*dt)/(dx**2)
print(F)

## Initial the temp
u = np.zeros((nx+1, nt+1))

#Initial t=0 profile
u[:, 0] = np.sin((np.pi*x)/L)

#boundary conditions
u[0,:]=u[-1,:]=0.0
for j in range(0, nt):
    for i in range(1, nx):
        u[i, j+1] = u[i, j] + F*(u[i+1, j]-2*u[i, j]+u[i-1, j])
print(u)
```

```python
for i in range(0, nt):
    if(i%10==0): plt.plot(x, u[:, i])
plt.xlabel('Position', fontsize=20)
plt.ylabel('Temperature', fontsize=20)
plt.show()
```
