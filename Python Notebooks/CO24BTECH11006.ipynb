```python
import numpy as np
import time
```

```python
r = 1
True_Area = np.pi * r**2
```

To find: Area of circle

Steps:
1. Make a square grid having size as the radius of circle and centred at the circle centre.
2. We can approximate the area of the circle by counting the number of smaller squares inside the circle.
3. Here, I am considering a square to be inside the circle if it's midpoint is inside the circle, i.e., if (x_mid^2 + y_mid^2 <= 1).
3. Then take the fraction of squares inside the circle by the total number of squares, and multiply fraction by the area of the square.
4. The resulting value will be the approximation of the area of a circle.

Inputs:

r, radius of circle

nx, number of division along x-directions

ny, number of division along y-directions

```python
nx = 1000
ny = 1000
x = np.linspace(-1, 1, nx+2)
y = np.linspace(-1, 1, ny+2)
count = 0
time_start = time.time()
for i in range(nx):
    for j in range(ny):
        xmid = 0.5*(x[i]+x[i+1])
        ymid = 0.5*(y[j]+y[j+1])
        if(np.sqrt(xmid**2+ymid**2)<=r): count+=1
Area_of_circle = (count/((nx)*(ny))) * (2*r)**2
time_end = time.time()
Error_G = Area_of_circle-True_Area
print(Area_of_circle, time_end-time_start, Error_G)
```

```python
r = 1
count = 0
iterations = 1000000
time_start = time.time()
for i in range(iterations):
    x = np.random.rand()*2*r - r
    y = np.random.rand()*2*r - r
    if(np.sqrt(x**2+y**2)<=r): count+=1
Area_of_circle = (count/iterations) * (2*r)**2
time_end = time.time()
Error_D = Area_of_circle-True_Area
print(Area_of_circle, time_end-time_start, Error_D)
```
