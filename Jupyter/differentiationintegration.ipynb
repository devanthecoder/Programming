```python
def f(x):
    return x**3
def delf(x):
    return 3*x**2
def fdelf(x, h):
    return (f(x+h)-f(x))/h
def bdelf(x, h):
    return (f(x)-f(x-h))/h
def cdelf(x, h):
    return (f(x+h)-f(x-h))/(2*h)
print(f(3))
print(delf(3) - fdelf(3, 0.01))
print(delf(3) - bdelf(3, 0.01))
print(delf(3) - cdelf(3, 0.01))
```

```python
import numpy as np
def TrapezoidalMethod(f, x1, x2, delx):
    arr = np.arange(x1, x2+delx, delx)
    farr = f(arr)
    area = (np.sum(farr[1:-1]) + (farr[0]+farr[-1])/2)*delx
    return area
f = lambda x: np.e**x
x1 = 0
x2 = 1
print(TrapezoidalMethod(f, x1, x2, 1e-5), np.e-1)
```
